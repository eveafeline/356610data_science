```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Define a function that takes a and return the value of the function


def f(a):
    return (a+1)**2+a-1

# You can use the following as well which is alberaically equivalent to the one above. 
#    return (a-1)*(a-1) - a/2 + 1
```

```python
# create an arrange starting at -0.75 going till 3.25 at interval of 0.1
x = np.arange(-6.20,3.25,0.1)
# for whole length of the array, call the function to fill up y
y = [f(k) for k in x]
```

```python
# Set axis labels and title. Show grid. 
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('y(x) = (x+1)^2 + x - 1')
plt.grid(True)

# Create and show the plot. 
plt.plot(x,y)
plt.show()
```

```python
# Create a function to return the derivative 
# You can use the either definition: The first one uses the derivative calculated 
# algebraically (2 v - 5/2) and the second uses the function with small increments
def derivative(v):
#     a = 2 * v - 5/2
    a = (f(v+0.01)-f(v-0.01))/0.02
    return a
```

```python
# Set up an array dydx (equal in size to x) to store derivative of the function
# dydx = np.zeros(len(x))
dydx = [derivative(k) for k in x]
# Populate array dydx by calling the derivative function incrementally 
# for i in range(len(x)):
#     dydx[i] = derivative(x[i])
```

```python
# Creating twin plot to show the function and derivative on the same graph

fig, plty = plt.subplots()
pltdx = plty.twinx()

plty.grid(True)
plty.set_ylabel('y(x)', color='r')
plty.tick_params('y', colors='r')
pltdx.set_ylabel('dy/dx', color='b')
pltdx.tick_params('y', colors='b')
plty.set_xlabel('x')
plty.set_title('y(x) = x^2 - 5x/2 + 2')

plty.plot(x,y,'r')
pltdx.plot(x, dydx, 'b')
plt.show()
```

```python
#generate a random number
xe = np.random.randn(1)
print(xe)
```

```python
# Let us define learning rate as alpha
alpha = 0.01
```

```python
# Set up number of iterations
iterations = 2500
```

```python
# set up a list to store evolving value of the minima
minima = np.zeros(iterations)
# put the inital (random) value of in the last cell of the list. 
# Note that in Python iteration[-1] will address the last cell of the list.
minima[iterations-1] = xe
# Now use the update formula to iterative improve your minima
for i in range(iterations):
    minima[i] = minima[i-1] - alpha * derivative(minima[i-1])
```

```python
plt.xlabel('iteration #')
plt.ylabel('minima')
plt.title('Gradient Descent')
plt.grid(True)
plt.plot(minima)
plt.show()
```
