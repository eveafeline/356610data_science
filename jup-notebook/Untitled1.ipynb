```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Define a function that takes a and return the value of the function

def f(a):
    return (a**2 - 5 * a /2 + 2)

# You can use the following as well which is alberaically equivalent to the one above. 
#    return (a-1)*(a-1) - a/2 + 1
```

```python
# create a range between -0.75 to 3.75 increment by 0.1
x = np.arange(-0.75, 3.25, 0.1)
# apply function to x to make y
y = [f(a) for a in x]
```

```python
# plot x / y
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('y(x) = x^2 - 5x/2 + 2')
plt.plot(x, y)
plt.grid(True)
plt.show()
```

```python
# create a derivative function
def derivative(v):
    return (f(v+0.01) - f(v-0.01)) / 0.02
```

```python
dydx = [derivative(a) for a in x]
```

```python
fig, plty = plt.subplots()
pltdx = plty.twinx()

plty.grid(True)
plty.set_ylabel('y(x)', color='r')
plty.tick_params('y', colors='r')
pltdx.set_ylabel('dy/dx', color='b')
pltdx.tick_params('y', colors='b')
plty.set_xlabel('x')
plty.set_title('y(x) = x^2 - 5x/2 + 2')

plty.plot(x,y,'r')
pltdx.plot(x, dydx, 'b')
plt.show()
```

```python
# find minimum
xe = np.random.randn(1)
xe
```

```python
alpha = 0.01
```

```python
minima = []
currX = xe
nextX = xe - alpha * derivative(xe)
while np.abs(nextX - currX) > 0.0000000001:
    currX = nextX
    nextX = currX - alpha * derivative(currX)
    minima.append(currX)
```
